Neste Notebook será analisado os documentos que foram classificados errados e tentar entender o porquê.

```python
# Importando Ferramentas Básicas

import pandas            as pd
import seaborn           as sns
import matplotlib.pyplot as plt
import re
```

- ***Num Atributos*** = 500

## Importação de Dados

```python
# Importando Dados

# DataFrame com textos, categoria, tokens e núm. de tokens
FP = pd.read_csv('False Positive_ComplementNB_500.csv')
FN = pd.read_csv('False Negative_ComplementNB_500.csv')
Total = pd.read_csv('noticias_final_alterado_limpo_500.csv')
```

```python
# Juntando os dados
Total['Classificação'] = ['Errada' if index in list(list(FN['Unnamed: 0']) + list(FP['Unnamed: 0'])) else 'Certa' for index in Total.index]

# Salvando DataFrame
Total.to_csv('Total_com_Class.csv')

# Visualizando modificações
Total.head()
```

## Análise Quantitativa

<br>

Vamos agora ver qual é a relação entre o número de tokens e classificação.

```python
sns.set_theme(style="darkgrid")

fig, ax = plt.subplots(figsize = (15,7.5))

ax = sns.kdeplot(data = Total, 
                  x = 'tamanho_lista_tokens',
                  hue = 'Classificação',
                  palette = {"Certa": "#33FFBD",  'Errada': "#FF5733"},
                  fill = True, 
                  common_norm = False, 
                  alpha = .5, 
                  linewidth = 1.5)

fig.suptitle('Distribuição de Tokens', fontsize = 25, fontweight = 'bold', y = 0.94)

plt.ylabel('Densidade', fontsize = 15, y = 0.5)
plt.xlabel('Número de Tokens', fontsize = 15, x = 0.5)

ax.set_xlim([-5, 50])

fig.savefig('Distribuicao_Tokens.png');
```

```python
def dic_values():
    
    dic = {}
    
    dic = {'0': ['Class. Certa',
                 round(Total.tamanho_lista_tokens[Total['Classificação'] == 'Certa'].mean(),2),
                 Total.tamanho_lista_tokens[Total['Classificação'] == 'Certa'].min(),
                 Total.tamanho_lista_tokens[Total['Classificação'] == 'Certa'].max()],
           '1': ['Class. Errada',
                 round(Total.tamanho_lista_tokens[Total['Classificação'] == 'Errada'].mean(),2),
                 Total.tamanho_lista_tokens[Total['Classificação'] == 'Errada'].min(),
                 Total.tamanho_lista_tokens[Total['Classificação'] == 'Errada'].max()]}
    
    df = pd.DataFrame.from_dict(dic,
                              orient = 'index',
                              columns = ['Tipo',
                                         'Média',
                                         'Mínimo', 
                                         'Máximo'])
    return df
```

## Análise Qualitativa

<br>

Aqui vamos analisar a correlação entre os que foram Classificados Corretamente ou não.

```python
# Dividindo os dados
Errada = Total[Total['Classificação'] == 'Errada']
Certa = Total[Total['Classificação'] == 'Certa']

# Visualizando a tabela
Certa.head()
```

Como a coluna lista_tokens está como string e não como lista, vamos desmembrá-la e criar uma lista com todos os tokens

```python
regex_pattern = "[^àÀáÁéÉçôõãúÚíÍ, A-Za-z0-9]+"

df_combinado = [Certa, Errada]

Lista_Certa = []
Lista_Errada = []

for df in df_combinado:
    for row in df.lista_tokens:           
        string_list = row.split(',')

    if len(string_list) == 1:
       string_list = re.sub(regex_pattern, "", string_list[0])

    else:
        string_list = [re.sub(regex_pattern, "", string) for string in string_list]
        
        if str(df) == 'Certa':
            Lista_Certa.extend(string_list)
            Lista_Certa = pd.Series(Lista_Certa)
            Tokens_Certos = Lista_Certa.value_counts()
            
        elif str(df) == 'Errada':
            Lista_Errada.extend(string_list)
            Lista_Errada = pd.Series(Lista_Errada)
            Tokens_Errados = Lista_Errada.value_counts()
```

Aqui vamos pegar o quantidade de occorências e dividir pelo número total de tokens únicos para ter um valor proporcional

```python
Prop_Errados = Tokens_Errados/len(Lista_Errada)
Prop_Certos = Tokens_Certos/len(Lista_Certa)
```

Aqui pegamos quais palavras só aparecem nas Corretas, quais só aparecem nas erradas e qual está em ambas

```python
Only_Correta = [word for word in list(Tokens_Certos.index) if word not in list(Tokens_Errados.index)]

Only_Errado = [word for word in list(Tokens_Errados.index) if word not in list(Tokens_Certos.index)]

Ambos = [word for word in list(Tokens_Certos.index) if word in list(Tokens_Errados.index)]
```

```python
# Ajeitando os Dataframes

Prop_Errados_df = pd.DataFrame(Prop_Errados).reset_index()
Prop_Certos_df = pd.DataFrame(Prop_Certos).reset_index()

Prop_Certos_df.rename(columns={"index": "Tokens_Certo", 0 : 'Value_Certo'}, inplace = True)
Prop_Errados_df.rename(columns={"index": "Tokens_Errado", 0 : 'Value_Errado'}, inplace = True)

# Visualizando mudanças
Prop_Certos_df.head()
```

```python
# Pegando os valores e as palavras que estão em ambos

df1 = Prop_Certos_df.loc[Prop_Certos_df['Tokens_Certo'].isin(Ambos)]
df2 = Prop_Errados_df.loc[Prop_Errados_df['Tokens_Errado'].isin(Ambos)]
```

```python
# Juntando os Dataframes em um
Tokens_Totais = df1.merge(df2, left_on= 'Tokens_Certo', right_on='Tokens_Errado')

# Organizando o Dataframe
Tokens_Totais.drop(labels = 'Tokens_Errado', axis = 1, inplace = True)
Tokens_Totais.rename(columns = {'Tokens_Certo' : 'Tokens em Comum'}, inplace = True)

# Calculando a diferença entre os valores
Tokens_Totais['Value_Diferença'] = Tokens_Totais.Value_Errado - Tokens_Totais.Value_Certo

# Colocando os valores em ordem decrescente
Tokens_Totais.sort_values(by = 'Value_Diferença', inplace = True, ascending = False)

# Visualizando mudanças
Tokens_Totais.tail()
```

```python
fig, (ax0, ax1) = plt.subplots( nrows = 2,
                                figsize = (15,20))

scatter1 = ax0.scatter(data = Tokens_Totais[:50],
                     x = 'Tokens em Comum',
                     y = 'Value_Diferença',
                     c = '#FF5733')

plt.subplots_adjust(hspace = 0.3)

scatter2 = ax1.scatter(data = Tokens_Totais[-50:],
                     x = 'Tokens em Comum',
                     y = 'Value_Diferença',
                     c = '#209F36')

fig.suptitle('Diferença na proporção de Tokens em Docs CC e CE', fontsize = 28, fontweight = 'bold', y = 0.94)

ax0.set_title('Tokens mais relevantes para classificar Erradamente (50 maiores Scores)',
              fontsize = 18,
              y = 1.03)

ax1.set_xlabel('Tokens',
               fontsize = 18,
               fontweight = 'bold')

ax0.set_ylabel('Densidade',
               fontsize = 16,
               fontweight = 'bold')

ax1.set_ylabel('Densidade',
               fontsize = 16,
               fontweight = 'bold')


ax1.set_title('Tokens mais relevantes para classificar Corretamente (50 menores Scores)',
              fontsize = 18,
              y = 1.03)


ax0.set_xticklabels(Tokens_Totais['Tokens em Comum'][:50], rotation=90)
ax1.set_xticklabels(Tokens_Totais['Tokens em Comum'][-50:], rotation=90);
```

```python
# Salvando um Dataframe com as palavras que só existem nas Certas, Erradas e em Ambas

# Criando um Dataframe com esses dados
df_tokens = pd.DataFrame()

df_tokens['Só Correta'] = pd.Series(Only_Correta)
df_tokens['Só Errada'] = pd.Series(Only_Errado)
df_tokens['Ambas'] = pd.Series(Tokens_Totais['Tokens em Comum'])

# Salvando Dataframe
df_tokens.to_csv('Tokens Análise FP FN.csv')
```
